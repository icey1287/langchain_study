```python
from pydantic import BaseModel, Field
from typing import List

class Aspect(BaseModel):
    """è¯„è®ºç»´åº¦"""
    name: str = Field(description="ç»´åº¦åç§°ï¼Œå¦‚ï¼šè´¨é‡ã€ä»·æ ¼ã€æœåŠ¡")
    score: int = Field(description="è¯„åˆ†ï¼Œ1-5")
    comment: str = Field(description="å…·ä½“è¯„ä»·")

class ProductReview(BaseModel):
    """äº§å“è¯„è®ºåˆ†æ"""
    overall_sentiment: str = Field(description="æ•´ä½“æƒ…æ„Ÿï¼špositive/negative/neutral")
    overall_score: int = Field(description="ç»¼åˆè¯„åˆ†ï¼Œ1-5")
    aspects: List[Aspect] = Field(description="å„ç»´åº¦è¯„ä»·")
    summary: str = Field(description="ä¸€å¥è¯æ€»ç»“")

# åˆ›å»ºç»“æ„åŒ–æ¨¡å‹
structured_model = model.with_structured_output(ProductReview, method="function_calling")
# æµ‹è¯•
review_text = """
è¿™æ¬¾ç¬”è®°æœ¬ç”µè„‘æ€§èƒ½éå¸¸å¼ºå¤§ï¼Œè¿è¡Œå¤§å‹è½¯ä»¶æ¯«æ— å‹åŠ›ã€‚
å±å¹•è‰²å½©é²œè‰³ï¼Œçœ‹è§†é¢‘å¾ˆèˆ’æœã€‚
ä¸è¿‡ä»·æ ¼æœ‰ç‚¹è´µï¼Œè€Œä¸”é£æ‰‡å™ªéŸ³è¾ƒå¤§ã€‚
å®¢æœæ€åº¦å¾ˆå¥½ï¼Œç‰©æµä¹Ÿå¿«ã€‚
æ€»ä½“æ¥è¯´è¿˜æ˜¯å€¼å¾—è´­ä¹°çš„ã€‚
"""

result = structured_model.invoke(
    f"åˆ†æä»¥ä¸‹äº§å“è¯„è®ºï¼š\n{review_text}"
)

print(f"æ•´ä½“æƒ…æ„Ÿ: {result.overall_sentiment}")
print(f"ç»¼åˆè¯„åˆ†: {result.overall_score}/5")
print(f"\nå„ç»´åº¦è¯„ä»·:")
for aspect in result.aspects:
    print(f"  - {aspect.name}: {aspect.score}/5 - {aspect.comment}")
print(f"\næ€»ç»“: {result.summary}")
```
